```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(
    {
        "subject": ["Segundo", "Valentino", "Doppio", "Alessandro", "Allan"],
        "cx_rebound": [174.701, 103.316, 157.888, 108.193, 188.973],
        "hc_rebound": [42.226, 36.078, 43.596, 41.053, 33.832],
        "cx_decline": [130.479, 79.956, 103.725, 71.808, 179.685],
        "hc_decline": [37.733, 22.705, 43.898, 50.438, 38.512],
        "spw_rate_rebound": [0.328, 0.254, 0.133, 0.327, 0.215],
        "spw_rate_decline": [0.304, 0.143, 0.167, 0.334, 0.462],
        "spw_amp_rebound": [0.00316351, 0.000808123, 0.00114461, 0.00117248, 0.000659866],
        "spw_amp_decline": [0.00167227, 0.00198515, 0.0016144, 0.00461481, 0.000890742]
    }
).set_index("subject")
print(df.to_markdown())
```

```python
print(df.corr(method='pearson').to_markdown())
```
